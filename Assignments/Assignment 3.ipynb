# Loading Data

```python
#Loading data
data1=pd.read_csv('C:\Wine-Quality-Prdiction\Wine-Quality-Prdiction\Data Set\winequality_red.csv')
data2=pd.read_csv('C:\Wine-Quality-Prdiction\Wine-Quality-Prdiction\Data Set\winequality_white.csv')
```

# Merging dataframes

```python
#Number of rows in first dataframe
length1=data1.shape[0]
length1
```

```python
#Number of rows in second datafram
length2=data2.shape[0]
length2
```

```python
#Additon of length of both rows
length=length1+length2
length
```

```python
#Addition of row in first dataframe
updated_index=np.arange(length2,length)
data1['updated_index']=updated_index
updated_index
data1
```

```python
#Saving previous index with name index in dataframe1
index=data1.index
data1['old_index']=index

#Concatinating first and second dataframe 
data1.set_index('updated_index', inplace = True)
data = pd.concat([data2,data1])
data
```

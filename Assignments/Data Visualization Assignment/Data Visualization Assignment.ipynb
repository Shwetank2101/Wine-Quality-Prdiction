## Q.1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),linestyle='-.',color='black')
plt.xlabel('X axis',size=25,color='Teal')
plt.ylabel('Y axis',size=25,color='Teal')
plt.title('Sine Wave Graph',size=30,color='Salmon')
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
list_1=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
arr1=[82,78,48,46,39]
arr2=[48,31,51,29,18]
arr3=[15,20,14,13,9]
var=np.arange(5)
width=0.30
plt.xticks(rotation='vertical')
plt.bar(list_1,arr1,width,label='Clothing',color='DeepSkyBlue')
plt.bar(var+width,arr2,width,label='Equipment',color='DeepPink')
plt.bar(var+width+width,arr3,width,label='Accessories',color='coral')
plt.grid(True)
plt.ylabel("No of Stores")
plt.xlabel("Cities")
plt.title('stores in each city')
plt.legend()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.barplot(x='survived',y='pclass',hue='alive',palette='gist_heat_r',data=titanic)
plt.ylabel("Passenger Class")
plt.xlabel("Survival of passengers")
plt.title('Alive passengers')
plt.legend()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
tips=sns.load_dataset('tips')
sns.stripplot( y="total_bill", data=tips,color='green')
plt.ylabel("No of Stores",size=15,color='RebeccaPurple')
plt.xlabel('X axis',size=15,color='RebeccaPurple')
plt.title('Total Bill',size=30,color='Salmon')
```

- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow
